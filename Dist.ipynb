```python
import pandas as pd
import numpy as np
from math import sqrt
import geopy.distance
from math import sin, cos, sqrt, atan2, radians
```

```python
def dist(lat1, long1, lat2, long2):
    
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(long1)
    lat2 = radians(lat2)
    lon2 = radians(long2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    
    return distance
```
